```python
import pandas as pd
```

```python
df = pd.read_csv('./framingham.csv')
df
```

```python
features = df.drop('TenYearCHD',axis='columns')
label = df.TenYearCHD
label
```

```python
features.columns[features.isna().any()]
```

```python
def impute_median(data):
    return data.fillna(data.median())
```

```python
features.glucose = features['glucose'].transform(impute_median)
features.education = features['education'].transform(impute_median)
features.heartRate = features['heartRate'].transform(impute_median)
features.totChol = features['totChol'].transform(impute_median)
features.BPMeds = features['BPMeds'].transform(impute_median)

# mengelompokkan berdasarkan atribut yang berhubungan dengan atribut lain

by_currentSmoker = features.groupby(['currentSmoker'])
features.cigsPerDay = by_currentSmoker['cigsPerDay'].transform(impute_median)

by_age = features.groupby(['male','age'])
features.BMI = by_age['BMI'].transform(impute_median)
```

```python
features.columns[features.isna().any()]
```

```python
from sklearn.model_selection import train_test_split

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.3)
```

```python
#Import Random Forest Model
from sklearn.ensemble import RandomForestClassifier

#Create a Gaussian Classifier
clf=RandomForestClassifier(n_estimators=100)

#Train the model using the training sets y_pred=clf.predict(X_test)
clf.fit(X_train,y_train)

y_pred=clf.predict(X_test)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Akurasi:",metrics.accuracy_score(y_test, y_pred))
```

```python
# normalisasi atribut menggunakan MinMaxScaler

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(features['glucose'])
features['glucose'] = scaler.transform(features['glucose'])

scaler.fit(features.education)
features.education = scaler.transform(features.education)

scaler.fit(features.heartRate)
features.heartRate = scaler.transform(features.heartRate)

scaler.fit(features.totChol)
features.totChol = scaler.transform(features.totChol)

scaler.fit(features.BPMeds)
features.BPMeds = scaler.transform(features.BPMeds)

scaler.fit(features.cigsPerDay)
features.cigsPerDay = scaler.transform(features.cigsPerDay)

scaler.fit(features.BMI)
features.BMI = scaler.transform(features.BMI)
```
