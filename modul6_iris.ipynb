## Naive Bayes

```python
# import dataset menggunakan library sklearn
import pandas as pd
from sklearn.datasets import load_iris
df = load_iris()
```

```python
df
```

```python
# memisahkan antara atribut data dengan atribut kelas
inputs = df.data
target = df.target
```

```python
# membuat data train dan data tes
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size = 0.4)
```

```python
X_train
```

```python
X_test
```

```python
# meminimalisasi skala nilai data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
# proses utama klasifikasi dengan algoritma naive bayes 
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
classifier.score(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test) 
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```
