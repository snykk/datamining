```python
import numpy as np 
import pandas as pd 
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import accuracy_score 
from sklearn.datasets import load_iris
from matplotlib import pyplot as plt
from sklearn import datasets
from sklearn import tree
```

```python
# 6.	Jelaskan kode pada baris 13  14
iris_data = load_iris()
print(type(iris_data))
for i in iris_data:
    print(i)
```

```python
iris=pd.DataFrame(iris_data.data)
print ("Features Name : ", iris_data.feature_names) 
print ("Dataset Shape: ", iris.shape) 
print ("Dataset: ",iris.head())
```

```python
print(iris.values[0])
```

```python
# 7.	Jelaskan kode pada baris 19  20
X = iris.values[:, 0:4] 
print(X)
```

```python
Y = iris_data.target
print(Y)
```

```python
# 8.	Jelaskan makna kode pada baris 22
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 100)
# print(X_train)
# print(X_test)
# print(y_train)
# print(y_test)
```

```python
# 9.	Jelaskan makna kode pada baris 24-25
clf= DecisionTreeClassifier(random_state = 100)
clf.fit(X_train, y_train)
print(clf)
```

```python
# 10.	Jelaskan makna kode pada baris 27-29
X=[[6.4,1.8 ,6.6 ,2.1]]
Y_pred=clf.predict(X)
print(Y_pred)
```

```python
# 11.	Jelaskan makna kode pada baris 30  31
Y_pred=clf.predict(X_test)
print(Y_pred)
# for i in Y_pred:
#     if i == 1:
#         print('versicolor')
```
