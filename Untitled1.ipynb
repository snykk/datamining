```python
#blok1
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
from sklearn.ensemble import BaggingClassifier
import pandas as pd
```

```python
df = pd.read_csv('datasetBST.csv')
df
```

```python
inputs = df.drop('bst',axis='columns')
targets = df.bst
```

```python
jk_dummies = pd.get_dummies(inputs['jenis kelamin'])
sp_dummies = pd.get_dummies(inputs['status pernikahan'])
inputs = pd.concat([inputs,jk_dummies,sp_dummies],axis='columns')
inputs.head(3)
```

```python
sp_dummies
```

```python
inputs.drop(['jenis kelamin','status pernikahan','P','tidak menikah'],axis='columns',inplace=True)
inputs.head()
```

```python
inputs
```

```python
# 1.	normalisasikan dataset iris menggunakan normalisasi minmax
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(inputs[['umur']])
inputs['umur'] = scaler.transform(inputs[['umur']])

scaler.fit(inputs[['anggota keluarga']])
inputs['anggota keluarga'] = scaler.transform(inputs[['anggota keluarga']])

scaler.fit(inputs[['jumlah tanggungan pendidikan']])
inputs['jumlah tanggungan pendidikan'] = scaler.transform(inputs[['jumlah tanggungan pendidikan']])

scaler.fit(inputs[['gaji']])
inputs['gaji'] = scaler.transform(inputs[['gaji']])


inputs.head()
```

```python
#blok 6
X_train, X_test, y_train, y_test = train_test_split(inputs, targets, test_size=0.4)
```

```python
from xgboost import XGBClassifier
model = XGBClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print('Forest Train accuracy %s' % model.score(X_train, y_train)) 
print('Trees Test accuracy %s' % accuracy_score(y_pred, y_test)) 
```

```python
from sklearn.metrics import confusion_matrix 
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0).fit(X_train, y_train)
y_predss = clf.predict(X_test)
clf.score(X_train, y_train)
```

```python
accuracy_score(y_predss, y_test)
```

```python
cm = confusion_matrix(y_test, y_predss)
cm
```
