## Naive Bayes

```python
# 1.	jelaskan attribut dataset titanic pada blok 2
#blok 1
import pandas as pd
df = pd.read_csv("adult.csv")
df.head()
```

```python
# 2.	jelaskan apakah tujuan dari kode blok 2
#blok 2
df.drop(['workclass','educational-num','marital-status','occupation','relationship','race','native-country'],axis='columns',inplace=True)
df.head()
```

```python
# 3.	jelaskan apakah tujuan dari kode blok 3
#blok 3
inputs = df.drop('income',axis='columns')
target = df.income
target
```

```python
inputs
```

```python
# 4.	jelaskan apakah tujuan dari kode blok 4 - 6
#blok 4
dummies = pd.get_dummies(inputs.gender)
dummies.head(3)
```

```python
#blok 5
inputs = pd.concat([inputs,dummies],axis='columns')
inputs.head(3)
```

```python
#blok 6
inputs.drop(['gender','Female'],axis='columns',inplace=True)
inputs.head()
```

```python
#blok 7
inputs.columns[inputs.isna().any()]
```

```python
#blok 10
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.3)
```

```python
# 7.	jelaskan tahapan klasifikasi yang dilakukan pada kode blok 11
#blok 11
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```
