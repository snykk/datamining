## Naive Bayes

```python
# 1.	jelaskan attribut dataset titanic pada blok 2
#blok 1
import pandas as pd
df = pd.read_csv("iniDsNaive.csv", delimiter='\t')
df.head()
```

```python
inputs = df.drop('KELAS',axis='columns')
label = df['KELAS']
```

```python
inputs
```

```python
#blok 10
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs,label,test_size=0.3)
```

```python
# 7.	jelaskan tahapan klasifikasi yang dilakukan pada kode blok 11
#blok 11
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
y_pred = model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
```
