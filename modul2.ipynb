### 1. Jelaskan makna dataset. Jelaskan pula makna data training dan data testing
#### data set
di machine learning data set adalah sekumpulan data yang nantinya dapat digunakam sebagai bahan mentah untuk diproses ke dalam sebuah algoritma machine learning entah untuk data training atau data testing. 
#### data training
sedangkan data training adalah data yang akan digunakan oleh machine learning dalam proses berlatih atau belajar sehingga dapat mengarah kepada hasil yang diharapkan. data -data ini disebut training set
#### data testing
Setelah proses training dilakukan pada sebuah algoritma Machine Learning, tahap selanjutnya adalah melakukan evaluasi terhadap performa algoritma tersebut atau biasa disebut testing. Pada proses testing, performa algoritma akan diuji menggunakan testing set, dimana testing set dan training set merupakan data yang berbeda.

```python
import csv
import numpy as np
import matplotlib.pyplot as plt

iris = []
with open('iris.csv') as csv_file:
    data = csv.reader(csv_file, delimiter='\t')
    for item in data:
        iris.append(item)

setosa = [iris[0]]
versicolor = [iris[0]]
verginica = [iris[0]]
unknown = [iris[0]]


for i in iris[1::]:
    if i[-1].lower() == " setosa":
        setosa.append(i)
    elif i[-1].lower() ==" versicolor":
        versicolor.append(i)
    elif i[-1].lower() ==" verginica":
        verginica.append(i)
    else:
        unknown.append(i)
    

print(setosa)
```

#### 8. Buatlah pararrel coordinat plot dari data yang berkelas setosa, versicolor, dan virginica. Berikan warna merah untuk setosa, hijau untuk versicolor, biru untuk virginica. Hint gunakan perintah plot dan operasi slicing pada numpy untuk mempermudah tugas anda 

```python
def kelas(nama_jenis):
    return [np.array(nama_jenis)[1::, 1].astype(np.float64),np.array(nama_jenis)[1::, 2].astype(np.float64),np.array(nama_jenis)[1::, 3].astype(np.float64),np.array(nama_jenis)[1::, 4].astype(np.float64)]

x = ['Petal Width', 'Petal Length', 'Sepal Width','Sepal Length']
y0 = kelas(setosa)
y1 = kelas(verginica)
y2 = kelas(versicolor)

plt.plot(x,y0,color='r',linestyle='-',label="Setosa")
plt.plot(x,y1,color='g',linestyle='-',label="Verginicaa")
plt.plot(x,y2,color='b',linestyle='-',label="Versicolor")

handles, labels = plt.gca().get_legend_handles_labels()
label = dict(zip(labels,handles))
plt.legend(label.values(),label.keys(), loc="lower right")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

#vectors to plot: 4D for this example
y1=[1,2.3,8.0,2.5]
y2=[1.5,1.7,2.2,2.9]

x = ['Petal Width', 'Petal Length', 'Sepal Width','Sepal Length']

fig,(ax,ax2,ax3) = plt.subplots(1, 3, sharey=False)

# plot the same on all the subplots
ax.plot(x,y1,'r-', x,y2,'b-')
ax2.plot(x,y1,'r-', x,y2,'b-')
ax3.plot(x,y1,'r-', x,y2,'b-')

# now zoom in each of the subplots 
ax.set_xlim([ x[0],x[1]])
ax2.set_xlim([ x[1],x[2]])
ax3.set_xlim([ x[2],x[3]])

# set the x axis ticks 
for axx,xx in zip([ax,ax2,ax3],x[:-1]):
    axx.xaxis.set_major_locator(ticker.FixedLocator([xx]))
ax3.xaxis.set_major_locator(ticker.FixedLocator([x[-2],x[-1]]))  # the last one

# EDIT: add the labels to the rightmost spine
for tick in ax3.yaxis.get_major_ticks():
    tick.label2On=True

# stack the subplots together
plt.subplots_adjust(wspace=0)

plt.show()
```
