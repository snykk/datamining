```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([7,10,11,14,17,17,19,20])
y = np.array([11,5,15,2,1,11,4,20])
y2 = np.array([a+1 for a in y])
y3 = np.array([a+2 for a in y])
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("latihan", fontdict = font1)
plt.xlabel("nilai  mendatar", fontdict = font2)
plt.ylabel("nilai vertikal", fontdict = font2)
plt.subplot(1,3,1)
plt.plot(x, y)
plt.subplot(1,3,2)
plt.plot(x, y2)
plt.subplot(1,3,3)
plt.plot(x, y3)
plt.show()
```
